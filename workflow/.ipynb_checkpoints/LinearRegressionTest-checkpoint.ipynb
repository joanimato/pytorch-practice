```python
import torch
from torch import nn 
import matplotlib.pyplot as plt
import numpy as np

torch.__version__
```

```python
# set up device
dev = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
#  create some noisy, but mostly linear data. 
weight = 0.5 
bias = 0.1 

noise = np.random.normal(0.1,0.05, 50)

X = torch.arange(0, 1, 0.02)
y = weight * X + bias + noise

# training and testing
tSplit = int(0.8 * len(X))
XTrain, yTrain = X[:tSplit].unsqueeze(1), y[:tSplit].unsqueeze(1)
XTest, yTest = X[tSplit:].unsqueeze(1), y[tSplit:].unsqueeze(1)
```

```python
print (len(XTrain), len(XTest))
```

```python
# visualize numbers
def plotPredictions(trainData=XTrain, trainLabels=yTrain, 
                    testData=XTest, testLabels=yTest, 
                    prediction=None):
    """
    Plots data and compares predictions
    """
    plt.figure(figsize=(8,6))
    
    plt.scatter(trainData, trainLabels, c="b", s=4, label="Training Data")
    plt.scatter(testData, testLabels, c="g", s=4, label="Testing Data")
    
    if prediction is not None:
        plt.scatter(testData, prediction, c="r", s=4 , label="Predictions")
    
    plt.legend(prop={"size": 14})
```

```python
plotPredictions()
```

```python
# building the linear model

class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linearLayer = nn.Linear(in_features=1, out_features=1)
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linearLayer(x)
    
```

```python
linearModel = LinearRegressionModel()
```

```python
# train model
linearModel.to(dev) # in case gpu is available
XTrain = XTrain.to(dev)
yTrain = yTrain.to(dev)
XTest = XTest.to(dev)
yTest = yTest.to(dev)

lossFn = nn.L1Loss()
optimizer = torch.optim.SGD(lr = 0.02, params=linearModel.parameters())

epochs = range(301)
for e in epochs:
    linearModel.train()
    
    yPred = linearModel(XTrain)
    
    loss = lossFn(yPred, yTrain)
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
    
    # testing
    
    linearModel.eval()
    
    with torch.inference_mode():
        testPrediction = linearModel(XTest)
        
        testLoss = lossFn(testPrediction, yTest)
    
    if e % 10 == 0:
        print(f"Epoch: {e}, Loss: {loss}, Test Loss: {testLoss}")
```

```python
with torch.inference_mode():
    y_preds = linearModel(XTest)
    
plotPredictions(prediction=y_preds.cpu())
```

```python
# saving the model 
from pathlib import Path

# create model directory
ModelPath = Path("models")
ModelPath.mkdir(parents=True, exist_ok=True)

# Create model save path

ModelName = "linearRegression.pth"
ModelSavePath = ModelPath / ModelName

# save the state dict

torch.save(linearModel, ModelSavePath)
```
